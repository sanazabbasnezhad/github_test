```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
data=pd.read_csv("Admission_Predict.csv")
```

```python
data.describe()
```

```python
data.drop(columns=["Serial No."])
```

```python
df=data
```

```python
df=df[(df["GRE Score"]<=340)|(df["GRE Score"]>=2601)]
df=df[(df["TOEFL Score"]<=120)|(df["TOEFL Score"]>=0)]
df=df[(df["CGPA"]<=10)|(df["CGPA"]>=0)]
df.dropna()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=np.asarray(df[["Serial No.","GRE Score","TOEFL Score","University Rating","SOP","CGPA","Research"]])
y=np.asarray(df["Chance of Admit "])
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.35)
```

```python
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
```

```python
x=np.linspace(0,1)
y=x
plt.scatter(y_test,y_pred)
plt.plot(x,y,color="red")
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import root_mean_squared_error
from sklearn.metrics import max_error
```

```python
print(r2_score(y_test,y_pred))
print(root_mean_squared_error(y_test,y_pred))
print(max_error(y_test,y_pred))
```

```python
from sklearn.preprocessing import PolynomialFeatures

from sklearn.pipeline import Pipeline
```

```python
steps=[("ploy_feature",PolynomialFeatures(degree=2)),("lr",LinearRegression(fit_intercept=False))]
model2= Pipeline(steps=steps)
model2.fit(x_train,y_train)
y_pred2=model2.predict(x_test)
```

```python
x=np.linspace(0,1,200)
y=x
plt.scatter(y_test,y_pred2)
plt.plot(x,y,color="black")
```

```python
print(r2_score(y_test,y_pred2))
print(root_mean_squared_error(y_test,y_pred2))
print(max_error(y_test,y_pred2))
#hi how
```
